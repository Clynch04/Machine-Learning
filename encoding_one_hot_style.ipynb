##### We will one-hot encode the country column from our basketball dataset.

#### Tasks:

##### Import the necessary library.
##### Build a one-hot encoder that uses a dtype of int and sparse=False. Name it one_hot_encoder.
##### Fit on X_column, transform it and save the results in an object named country_encoded.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

##### Loading in the data

```python
bball_df = pd.read_csv('bball.csv')
bball_df = bball_df[(bball_df['position'] =='G') | (bball_df['position'] =='F')]
```

##### Split the dataset

```python
df_train, df_test = train_test_split(bball_df, test_size=0.2, random_state=7)

X_column = df_train[['country']]
```

##### Build the tranformer and fit on it

Build a one-hot encoder that uses a dtype of int and sparse=False. Name it one_hot_encoder.

Fit on X_column

```python
one_hot_encoder = OneHotEncoder(dtype=int, sparse_output=False)
one_hot_encoder.fit(X_column)
```

##### Transform the column country

transform it and save the results in an object named country_encoded.

```python
country_encoded = one_hot_encoder.transform(X_column)
```

##### Print the output of country_encoded

```python
print(country_encoded)
```

##### Let's look at this in a dataframe

```python
pd.DataFrame(data=country_encoded, columns=one_hot_encoder.get_feature_names_out(['country']), index=X_column.index).head()
```
